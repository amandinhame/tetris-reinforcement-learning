### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from learner import *
```

### 2500 trials - linear decreasing exploration factor

#### Training data

```python
# Preparing data
lin_2500_train = pd.read_csv(os.path.join("lin_2500", "training.csv"), header = None)
lin_2500_train['holes_mean_window'] = (lin_2500_train[1]).rolling(window = 100, center = False).mean()
lin_2500_train['lines_mean_window'] = (lin_2500_train[2]).rolling(window = 100, center = False).mean()

# Plotting charts
# Holes and lines
plt.plot(lin_2500_train[0], lin_2500_train['holes_mean_window'], label='holes')
plt.plot(lin_2500_train[0], lin_2500_train['lines_mean_window'], label='lines')
plt.xlabel('Trials')
plt.ylabel('Quantity')
plt.title("2500 Trial Training - Linear - Training\n100-Trial rolling mean of holes and lines")
plt.legend()
plt.show()

# Exploration factor
plt.plot(lin_2500_train[0], lin_2500_train[3], label = 'epsilon')
plt.xlabel('Trials')
plt.ylabel('Parameter value')
plt.title("2500 Trial Training - Linear - Training\nExploration factor")
plt.legend()
plt.show()
```

#### Random actions test

```python
# Preparing data
lin_2500_rand = pd.read_csv(os.path.join("lin_2500", "testing_random.csv"), header = None)

# Plotting charts
# Holes, lines and success
plt.plot(lin_2500_rand[0], lin_2500_rand[1], label='holes')
plt.plot(lin_2500_rand[0], lin_2500_rand[2], label='lines')
plt.plot(lin_2500_rand[0], lin_2500_rand[3], label='success')
plt.xlabel('Trials')
plt.ylabel('Quantity')
plt.title("2500 Trial Training - Linear - Random Test\nNumber of holes, lines and success")
plt.legend()
plt.show()

# summary
lin_2500_rand.columns = ['Trials','Holes','Lines','Success']
lin_2500_rand.describe()
```

#### Learning test

```python
# Preparing data
lin_2500_test = pd.read_csv(os.path.join("lin_2500", "testing.csv"), header = None)

# Plotting charts
# Holes, lines and success
plt.plot(lin_2500_test[0], lin_2500_test[1], label='holes')
plt.plot(lin_2500_test[0], lin_2500_test[2], label='lines')
plt.plot(lin_2500_test[0], lin_2500_test[3], label='success')
plt.xlabel('Trials')
plt.ylabel('Quantity')
plt.title("2500 Trial Training - Linear - Learning Test\nNumber of holes, lines and success")
plt.legend()
plt.show()

# summary
lin_2500_test.columns = ['Trials','Holes','Lines','Success']
lin_2500_test.describe()
```
